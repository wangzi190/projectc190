```java
import java.util.Scanner;
import java.lang.Math;

public class Menu {
    public final String[] kinematics = {
        "vf = vi + a*t",
        "x = vi*t + 0.5a*t^2",
        "x = 0.5(vi + vf)t",
        "(vf)^2 = (vi)^2 + 2ax"
    };
    public Menu() {
        Scanner sc = new Scanner(System.in);
        
        this.print();
        boolean quit = false;
        while (!quit) {
            try {
                int choice = sc.nextInt();
                //System.out.print("" + choice + ": ");
                quit = this.action(choice);
            } catch (Exception e) {
                sc.nextLine();
                System.out.println(e + ": Not a number, try again.");
            }
        }
        sc.close();
    }

    private void print() {
        System.out.println("-------------------------\n");
        System.out.println("Select any of the following:");
        System.out.println("-------------------------\n");
        System.out.println("0 - Print the kinematic equations");
        System.out.println("1 - Generate and solve a math problem");
        System.out.println("2 - Roll a 6-sided die and print the result");
        System.out.println("3 - Quit");
        System.out.println("-------------------------\n");
    }

    private boolean action(int selection) {
        boolean quit = false;

        switch (selection) {
            case 0:
                for (int i = 0; i < kinematics.length; i++) {
                    System.out.println(kinematics[i]);
                }
            break;

            case 1:
                int x = (int)(Math.random() * 101);
                int y = (int)(Math.random() * 101);
                int operator = (int)(Math.random() * 2);
                int answer;
                if (operator == 0) {
                    answer = x + y;
                    System.out.println("Your generated math problem is " + x + " + " + y + ".");
                    System.out.println("The answer to the problem is " + answer + ".");
                } else if (operator == 1) {
                    if (x > y) {
                        answer = x - y;
                        System.out.println("Your generated math problem is " + x + " - " + y + ".");
                    } else {
                        answer = y - x;
                        System.out.println("Your generated math problem is " + y + " - " + x + ".");
                    }
                    System.out.println("The answer to the problem is " + answer + ".");
                }
            break;

            case 2:
                int roll = (int)(Math.random() * 7);
                System.out.println("You roll the 6-sided die, and get " + roll + "!");
            break;

            case 3:
                System.out.println("Quitting program..");
                quit = true;
            break;
                    
            default:
                System.out.println("Invalid choice. Please try again.");
        }
        return quit;
    }

    static public void main(String[] args)  {  
        new Menu();
    }

}
Menu.main(null);
```
