```javascript
var tape = [];
var pointer = 0;
var limit = 50; // <--how many spaces of tape to allocate to the machine

function print(symbol, position) {
    tape[position] = symbol;
};

function erase(position) {
    tape[position] = '';
};

function b() {
    console.log('b');
    for (let i = 0; i < 2; i++) {
        print('@', pointer);
        pointer++;
    }
    print('0', pointer);
    pointer = pointer + 2;
    print('0', pointer);
    pointer = pointer - 2;
    o(tape[pointer]);
};

function o(symbol) {
    console.log('o' + ', ' + pointer + ', ' + symbol);
    if (symbol=='1') {
        pointer++;
        print('x', pointer);
        pointer = pointer - 3;
        if (tape.length > limit) {
            throw new Error("Calculation complete.");
        } else {
            o(tape[pointer]);
        }
    } else if (symbol=='0') {
        q(tape[pointer]);
    }
};

function q(symbol) {
    console.log('q' + ', ' + pointer + ', ' + symbol);
    if (symbol==('0' || '1')) {
        pointer = pointer + 2;
        if (tape.length > limit) {
            throw new Error("Calculation complete.");
        } else {
            q(tape[pointer]);
        }
    } else if (!symbol) {
        print('1', pointer);
        pointer--;
        p(tape[pointer]);
    }
};

function p(symbol) {
    console.log('p' + ', ' + pointer + ', ' + symbol);
    if (symbol=='x') {
        erase(pointer);
        pointer++;
        q(tape[pointer]);
    } else if (symbol=='@') {
        pointer++;
        f(tape[pointer]);
    } else if (!symbol) {
        pointer = pointer - 2;
        if (tape.length > limit) {
            throw new Error("Calculation complete.");
        } else {
            p(tape[pointer]);
        }
    }
};

function f(symbol) {
    console.log('f' + ', ' + pointer + ', ' + symbol);
    if (!symbol) {
        print('0', pointer);
        pointer = pointer - 2;
        o(tape[pointer]);
    } else {
        pointer = pointer + 2;
        if (tape.length > limit) {
            throw new Error("Calculation complete.");
        } else {
            f(tape[pointer]);
        }
    }
};

b();

console.log(tape);
```
